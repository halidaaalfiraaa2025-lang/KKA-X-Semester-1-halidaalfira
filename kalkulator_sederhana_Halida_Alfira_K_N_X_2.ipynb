```python

jari_jari = int(input("Jari jari tabung adalah (cm): "))
tinggi = int(input("Tinggi tabung adalah (cm): "))

volume = 22/7 * jari_jari * jari_jari * tinggi

print(f"\n Volume Tabung yaitu {volume} cm^3")
```

```python
panjang = int(input("Masukkan panjang dari persegi panjang (cm): "))
lebar = int(input("Masukkan lebar dari persegi panjang (cm): "))

luas = panjang * lebar

print(f"\nLuas dari persegi panjang yaitu {luas} cm^2")
```
